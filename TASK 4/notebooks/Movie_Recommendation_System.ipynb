# movies['cast'][0] -> Gives the first movie's cast.

# ast.literal_eval(movies['cast'][0])[0]['name'] -> Gets the value associated with the "name" key.

# Now we will create our first convert() function

```python
def convert(text):
    result = []

    for item in ast.literal_eval(text):
        result.append(item['name'])
    return result
```

# Applying convert() function

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

# Creating convert3() Function

```python
# Through this function, we are storing only top three cast members
```

```python
def convert3(text):
    count = 0
    result = []

    for actor in ast.literal_eval(text):
        if(count<3):
            result.append(actor['name'])
            count+=1
        else:
            break
    return result
```

# Applying convert3() function

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies[['title', 'cast']].head()
```

```python
def fetch_director(text):
    result = []
    
    for crew in ast.literal_eval(text):
        if(crew['job'] == 'Director'):
            result.append(crew['name'])

    return result
    
```

# Applying fetch_director() function

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies[['title', 'crew']].head()
```

# Removing Spaces from Names

```python
# Creating our helper function
```

```python
def collapse(text):
    result=[]

    for item in text:
        result.append(item.replace(" ", ""))
    return result
```

```python
# TEST 1

collapse(['Sam Worthington', 'Zoe Saldana'])
```

# Applying helper() function on dataset

```python
movies['cast'] = movies['cast'].apply(collapse)

movies['crew'] = movies['crew'].apply(collapse)

movies['genres'] = movies['genres'].apply(collapse)

movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

# Now we will create a 'Tags' Column
# This column contains:
# - Overview

# Final Data Preprocessing

```python
# Keeping only required columns

new_df = movies[['movie_id', 'title', 'tags']].copy()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

# Text Vectorization using CountVectorizer

# Importing PorterStemmer

```python
from nltk.stem.porter import PorterStemmer
```

```python
# Creating an object

ps = PorterStemmer()
```

# Creating Stem function

```python
def stem(text):
    result = []

    for word in text.split():
        result.append(ps.stem(word))

    return " ".join(result)
```

```python
stem("loved loving loves")
```

# Applying Stem function

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

# Importing CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

# Creating Vectorizer

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

# Importing Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0][0]
```

# Building the Actual Recommendation System

```python
def recommend(movie):

    movie = movie.lower()
    
    matching_movies = new_df[new_df['title'].str.lower() == movie]

    if matching_movies.empty :
        print("Movie not found.")
        return

    movie_index = matching_movies.index[0]

    distances = similarity[movie_index]
    
    # Sort movies by similarity score
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])

    print("\nRecommended Movies:\n")

    # Print top 5 recommendations
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title)

    list(enumerate(distances))
```

# Testing the system

```python
recommend('avatar')
```

```python
recommend('The Dark Knight')
```

```python
recommend('Toy Story')
```
